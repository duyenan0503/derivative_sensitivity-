```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
r = 0.03
vol = 0.18
strike_price = 285
maturities = [0.1, 1, 2, 5, 10]
colors = ['red', 'blue', 'green', 'purple', 'orange']
x = np.arange(0, 600, 0.01)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
for i, maturity in enumerate(maturities):
    d1_call = (np.log(x / strike_price) + (r + vol**2 / 2) * maturity) / (vol * np.sqrt(maturity))
    delta_call = norm.cdf(d1_call)
    d1_put = (np.log(x / strike_price) - (r + vol**2 / 2) * maturity) / (vol * np.sqrt(maturity))
    delta_put = norm.cdf(d1_put) - 1
    total_delta = delta_call + delta_put
    ax1.plot(x, total_delta, label=f'Maturity: {maturity} years', color=colors[i])
ax1.set_title('Delta of Straddle at Different Maturities')
ax1.set_xlabel('Underlying Spot Price')
ax1.set_ylabel('Delta')
ax1.axhline(0, color='black', lw=1, linestyle='--')
ax1.legend(loc='upper left')
ax1.grid()
for i, maturity in enumerate(maturities):
    gamma_call = norm.pdf(d1_call) / (x * vol * np.sqrt(maturity))
    gamma_put = norm.pdf(d1_put) / (x * vol * np.sqrt(maturity))
    total_gamma = gamma_call + gamma_put
    ax2.plot(x, total_gamma, label=f'Maturity: {maturity} years', color=colors[i])
ax2.set_title('Gamma of Straddle at Different Maturities')
ax2.set_xlabel('Underlying Spot Price')
ax2.set_ylabel('Gamma')
ax2.axhline(0, color='black', lw=1, linestyle='--')
ax2.legend(loc='upper left')
ax2.grid()
for i, maturity in enumerate(maturities):
    vega_call = x * norm.pdf(d1_call) * np.sqrt(maturity)
    vega_put = x * norm.pdf(d1_put) * np.sqrt(maturity)
    total_vega = vega_call + vega_put
    ax3.plot(x, total_vega, label=f'Maturity: {maturity} years', color=colors[i])
ax3.set_title('Vega of Straddle at Different Maturities')
ax3.set_xlabel('Underlying Spot Price')
ax3.set_ylabel('Vega')
ax3.axhline(0, color='black', lw=1, linestyle='--')
ax3.legend(loc='upper left')
ax3.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def bs_vega(S, K, r, q, sigma, T):
    d1 = (np.log(S / K) + (r - q + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    return S * np.exp(-q * T) * norm.pdf(d1) * np.sqrt(T)

def price_KI_MC(S0, K, B, r, q, sigma, T, steps=200, paths=200000):
    dt = T / steps
    drift = (r - q - 0.5 * sigma**2) * dt
    vol = sigma * np.sqrt(dt)

    S = S0 * np.ones(paths)
    hit = np.zeros(paths, dtype=bool)

    for _ in range(steps):
        S = S * np.exp(drift + vol * np.random.randn(paths))
        hit = hit | (S >= B)

    payoff = np.where(hit, np.maximum(S - K, 0.0), 0.0)
    return np.exp(-r*T) * payoff.mean()

def ki_vega(S0, K, B, r, q, sigma, T, bump=0.01):
    p_up = price_KI_MC(S0, K, B, r, q, sigma + bump, T)
    p_down = price_KI_MC(S0, K, B, r, q, sigma - bump, T)
    return (p_up - p_down) / (2*bump)

S0 = 100
K = 90
r = 0.01
q = 0.00
sigma = 0.25
T = 1.0

barriers = np.linspace(101, 120, 12)

vanilla_vegas = []
ki_vegas = []
ko_vegas = []

for B in barriers:
    vv = bs_vega(S0, K, r, q, sigma, T)
    kv = ki_vega(S0, K, B, r, q, sigma, T)
    kov = vv - kv

    vanilla_vegas.append(vv)
    ki_vegas.append(kv)
    ko_vegas.append(kov)

plt.figure(figsize=(10,6))
plt.plot(barriers, vanilla_vegas, label="Vanilla Vega", linewidth=2)
plt.plot(barriers, ki_vegas, label="Knock-In Vega", linewidth=2)
plt.plot(barriers, ko_vegas, label="Knock-Out Vega", linewidth=2, linestyle="--")
plt.xlabel("Barrier Level")
plt.ylabel("Vega")
plt.title("Vega of Vanilla vs Knock-In vs Knock-Out Call Option")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
r = 0.03
vol = 0.18
strike_buy = 285
strike_sell = 260
x = np.linspace(6, 600, 1000)
maturities = [0.1, 1, 2, 5, 10]
payoff_buy = 2 * np.maximum(x - strike_buy, 0)
payoff_sell = np.maximum(x - strike_sell, 0)
payoff_total = payoff_buy - payoff_sell
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(x, payoff_total, color='red', label='Total Payoff (2x1 Ratio Spread)', linewidth=2)
ax1.axhline(0, color='black', lw=1, linestyle='--')
ax1.set_xlabel('Underlying Spot Price')
ax1.set_ylabel('Payoff', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax2 = ax1.twinx()
for maturity in maturities:
    d1_buy = (np.log(x / strike_buy) + (r + vol**2 / 2) * maturity) / (vol * np.sqrt(maturity))
    d1_sell = (np.log(x / strike_sell) + (r + vol**2 / 2) * maturity) / (vol * np.sqrt(maturity))
    delta_buy = norm.cdf(d1_buy)
    delta_sell = norm.cdf(d1_sell)
    delta_portfolio = delta_buy - 2 * delta_sell
    ax2.plot(x, delta_portfolio, label=f'Delta (Maturity: {maturity} years)')
ax1.set_title('Payoff and Delta of 2x1 Call Ratio Spread vs. Underlying Spot Price')
ax2.set_ylabel('Delta', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.grid()
plt.axvline(260, color = 'black')
plt.axvline(285, color = 'black')
plt.tight_layout()
plt.show()
```

```python
def black_scholes(isCall, S, K, r, q, T, sig):
  d1 = (np.log(S / K) + (r - q + 0.5 * sig**2) * T) / (sig * np.sqrt(T))
  d2 = d1 - sig * np.sqrt(T)
  if isCall:
    option_price = (S * np.exp(-q * T) * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2))
  else:
    option_price = (K * np.exp(-r * T) * norm.cdf(-d2) - S * np.exp(-q * T) * norm.cdf(-d1))
  return option_price
S = float(input("Enter the current spot price: "))
K = float(input("Enter the strike price: "))
r = float(input("Enter the risk-free price: "))
q = float(input("Enter the dividend yield price: "))
T = float(input("Enter the time to maturity: "))
sig = float(input("Enter the volatility: "))
call_price = black_scholes(True, S, K, r, q, T, sig)
print("The price of call option is: ", call_price)
def implied_by_bisection(S, K, r, q, T, sig, tol=1e-5, is_call=True):
    target_price = black_scholes(S, K, r, q, T, sig, is_call)
    low_vol, high_vol = 0.0001, 5
    target_vol = None

    while True:
        middle_vol = (low_vol + high_vol) / 2
        middle_price = black_scholes(S, K, r, q, T, middle_vol, is_call)

        if abs(middle_price - target_price) < tol:
            target_vol = middle_vol
            break
        elif middle_price < target_price:
            low_vol = middle_vol
        else:
            high_vol = middle_vol
        if high_vol - low_vol < tol:
            target_vol = middle_vol
            break

    return target_vol

implied_vol = implied_by_bisection(S, K, r, q, T, sig)
print("The implied volatility, using Bisection, is:", implied_vol)
```
